```python
import numpy as np
xi1 = np.array([30, 50, 70 ,80 ,100])
xi2 = np.array([0, 1, 1, 2, 1])
yi = np.array([0, 1, 0, 1, 1])
```

```python
import matplotlib.pyplot as plt
from matplotlib import patches
%pylab inline
pylab.rcParams['figure.figsize'] = 15, 5
```

```python
donateY = np.where(yi == 1)[0]
donateN = np.where(yi == 0)[0]
```

```python
plt.plot(xi1[donateY], xi2[donateY], 'g.', label='Donated')
plt.plot(xi1[donateN], xi2[donateN], 'r.', label='Did not donate')
plt.xlabel('Income')
plt.ylabel('Num Websites')
plt.legend()
ax = plt.gca()
ax.add_patch(patches.Polygon([[25,-1], [25, 1], [75,-1]], color = 'red', fill=False))
plt.savefig('figure.png')
plt.show()
```
